# Calculating new threshold

```python
df_val_all_models[['mes',
                   'es_fraude',
                   'M1_predictions',
                   'M2_predictions',
                   'M3_predictions',
                   'M4_predictions',
                   'M5_predictions']]
```

```python
from sklearn.metrics import (
    f1_score,
    precision_score,
    recall_score,
    confusion_matrix
)
```

```python
df = df_val_all_models[['mes',
                   'es_fraude',
                   'M1_predictions',
                   'M2_predictions',
                   'M3_predictions',
                   'M4_predictions',
                   'M5_predictions']]
```

```python
# error cost
FP_COST = 30_000
FN_COST = 1_750_000

prediction_cols = [c for c in df.columns if c.endswith('_predictions')]

y_true = df['es_fraude'].astype(int)

thresholds = np.arange(0, 1.01, 0.01)

for model_col in prediction_cols:

    results = []

    y_score = df[model_col]

    for threshold in thresholds:

        y_pred = (y_score >= threshold).astype(int)

        tn, fp, fn, tp = confusion_matrix(
            y_true,
            y_pred,
            labels=[0, 1]
        ).ravel()

        precision = precision_score(
            y_true,
            y_pred,
            zero_division=0
        )

        recall = recall_score(
            y_true,
            y_pred,
            zero_division=0
        )

        f1 = f1_score(
            y_true,
            y_pred,
            zero_division=0
        )

        risk = fp * FP_COST + fn * FN_COST

        results.append({
            'threshold': threshold,
            'f1': f1,
            'precision': precision,
            'recall': recall,
            'risk': risk,
            'fp': fp,
            'fn': fn,
            'tp': tp,
            'tn': tn
        })

    results_df = pd.DataFrame(results)

    # metrics
    plt.figure(figsize=(12, 6))

    plt.plot(results_df['threshold'], results_df['f1'], label='F1')
    plt.plot(results_df['threshold'], results_df['precision'], label='Precision')
    plt.plot(results_df['threshold'], results_df['recall'], label='Recall')

    plt.xlabel('Threshold')
    plt.ylabel('Metric value')
    plt.title(f'{model_col} - Classification metrics')
    plt.grid(True)
    plt.legend()

    plt.show()

    # finance risc
    plt.figure(figsize=(12, 6))

    plt.plot(results_df['threshold'], results_df['risk'])

    best_threshold = results_df.loc[
        results_df['risk'].idxmin(),
        'threshold'
    ]

    best_risk = results_df['risk'].min()

    plt.axvline(
        best_threshold,
        linestyle='--',
        label=f'Best threshold={best_threshold:.2f}'
    )

    plt.xlabel('Threshold')
    plt.ylabel('Risk')
    plt.title(
        f'{model_col} - Financial risk\n'
        f'Min risk={best_risk:,.0f}'
    )

    plt.grid(True)
    plt.legend()

    plt.show()

    print(
        f'\n{model_col}\n'
        f'Best threshold by risk: {best_threshold:.2f}\n'
        f'Min risk: {best_risk:,.0f}'
    )
```

```python
prediction_cols = [c for c in df.columns if c.endswith('_predictions')]

df_binary = df.copy()

df_binary[prediction_cols] = (
    df_binary[prediction_cols] >= 0.01
).astype(int)

df_binary.to_csv('df_preds_new_threshold.csv')
```
